## Te Waves in a Rectangular Wave Guide
### Separation of variables Method
for TE waves
$$(\nabla ^2 +(\frac{\omega }{c})^2-k^2)B_z = 0$$
separate variables:
$$\frac{1}{X}\frac{d^2 X}{dx^2} = -k_x^2 $$
$$\frac{1}{Y}\frac{d^2 Y}{dy^2} = -k_y^2 $$
solve(for TE):
$$B_z = B_0 cos(m \pi x/a) cos(n \pi y/a)$$

```python
# the plots of different mode of TE waves in a rectangular wave guide
import numpy as np
import matplotlib.pyplot as plt
```

```python
# m,n reprents the different modes
# assume B_z=1,a=1
from turtle import mode


def Bz(m, n, x):
    return np.cos(m * np.pi * x) * np.cos(n * np.pi * x)


fig, ax = plt.subplots()
X = np.linspace(0, 1, 100)

mode10 = Bz(1, 0, X)
mode11 = Bz(1, 1, X)
mode20 = Bz(2, 0, X)
mode21 = Bz(2, 1, X)
mode22 = Bz(2, 2, X)

ax.plot(X, mode10, label="m=1,n=0")
ax.plot(X, mode11, label="m=1,n=1")
ax.plot(X, mode20, label="m=2,n=0")
ax.plot(X, mode21, label="m=2,n=1")
ax.plot(X, mode22, label="m=2,n=2")
ax.legend()
ax.set_xlabel("x")
ax.set_ylabel("Bz(x)")

plt.savefig("waveguide.svg")
plt.show()
```
